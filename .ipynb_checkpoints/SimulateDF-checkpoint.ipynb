```python
import pandas as pd
import numpy as np
import random
```

```python
def get_unique_category_values(v):
    return list(set( v ) )

def get_random_categories( items , n  ):
    return random.choices(items, k=n)


def get_random_numbers( min , max, n ):
    values = []
    for i in range(n):
        values.append(
            random.randint(min, max)
        )
    return values

def get_random_dates( min , max , n ):
    items = pd.date_range( start = min , end = max , freq = 'D' )
    return get_random_categories( items , n )

def get_dataframe( sett , num_records ):
    metadata_df = {}

    types = [ 'categorico' , 'numerico' , 'date' ]
    
    for k in sett.keys():
        if sett[k]["tipo"] in types:
            if sett[k]['tipo'] == "categorico":
                values = get_random_categories( sett[k]['conf'] , num_records  )

            if sett[k]['tipo'] == "numerico":
                values = get_random_numbers( 
                    sett[k]['conf']['min'] 
                    , sett[k]['conf']['max']
                    , num_records 
                )
        
            if sett[k]['tipo'] == "date":
                values = get_random_dates( 
                    sett[k]['conf'][0]            
                    , sett[k]['conf'][1] 
                    , num_records 
                )
                
            
            metadata_df[k] = values
            
    return pd.DataFrame( metadata_df )
```

```python
def get_settins(df):

    settings = {}
    
    for c in df.columns:
        if df[c].dtypes == "object":
            settings[c] =  {
                "tipo" : "categorico"
                , "conf": get_unique_category_values( 
                        df[c] 
                    )
            }
    
        if df[c].dtypes in  ["int64","float64"]:
            settings[c] =  {
                "tipo" : "numerico"
                , "conf":{
                    "min": np.min( df[c] )
                    , "max": np.min( df[c] )
                }
            }
    
        if df[c].dtypes == "datetime64[ns]":
            settings[c] =  {
                "tipo" : "date"
                , "conf":[ 
                    np.min( df[c] )
                    , np.max( df[c] ) 
                ]
            }
    return settings
        
```

```python
settings  = get_settins(sales_v1)
```

```python
df = get_dataframe( settings , 10000 )
```
